```python
from bs4 import BeautifulSoup
```

```python
doc=BeautifulSoup(response.text)
```

```python
id1=doc.find("span",attras={"id":"id1"})
```

```python
table =doc.table
```

```python
author_graph =graphviz.Graph()
tbody=table.tbody
for tr in tbody.find_all("tr"):
    tds=tr.find_all("td")
    if len(tds) < 4 :
        continue
    authors=tds[3].text
    splitted=authors.split(",")
    for combo in itertools.combinations(splitted,2):
        author_graph.edge(combo[0],combo[1])
author_graph.save("author_graph.dot")
author_graph.render()
```

```python
author_dict={}
for tr in doc.table.tbody.find_all("tr"):
    tds=tr.find_all("td")
    if len(tds) < 4 :
        continue
    pep_id =tds[1].text
    authors=tds[3].text
    author_list=authors.split(",")
    for author in author_list:
        if author in author_dict:
            author_dict[author].append(pep_id)
        else:
            author_dict[author] = [pep_id]
```

```python
author_dict
```

```python
max_num=0
max_author=""
for author in author_dict:
    if len(author_dict[author]) > max_num:
        max_num=len(author_dict[author])
        max_author =author
```

```python
import requests
```

```python
res = requests.get("https://peps.python.org/api/peps.json")
```

```python
print(res.text)
```

```python
handle = open("pep.txt","w")
handle.write(res.text)
handle.close()
```

```python
import json
```

```python
pep_dict=json.loads(res.text)
```

```python
list(pep_dict.keys())
```

```python
pep_dict["2"]["authors"]
```

```python
accepted = []
for pep_id in pep_dict:
    if "status" in pep_dict[pep_id] and pep_dict[pep_id]["status"]=="Accepted":
        accepted.append(pep_id)
```

```python
accepted
```

```python
list(filter(lambda x:"status" in pep_dict[pep_id] and pep_dict[pep_id]["status"]=="Accepted",pep_dict))
```

```python
with open("pep.json","w") as f:
    json.dump(pep_dict,f)
```

```python
import csv
```

```python
header= ["ID" ,"title" ,"authors"]
with open("pep.csv","w",encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    for ID in pep_dict:
        writer.writerow([ID,pep_dict[ID]["title"],pep_dict[ID]["authors"]])
```

```python
with open("pep.csv","r",encoding="utf-8") as f:
    reader = csv.reader(f)
    content = list(reader)
```
